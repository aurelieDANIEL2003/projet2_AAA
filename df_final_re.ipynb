```python
import pandas as pd
import numpy as np
```

```python
chemin_bd = r"./bd_ignore/"
```

```python
df_top_5_actors_per_periods = pd.read_csv(chemin_bd + 'resultat/df_top_5_actors_per_periodsa.csv')
```

```python
df_top_5_actors_per_periods.head(10)
```

```python
df_top_5_actors_per_periods.info()
```

```python
df_names = pd.read_csv(chemin_bd + 'name.basics.tsv', sep='\t')  # Dataset des acteurs
df_names.head()
```

```python
df_names_titres = df_names.copy()
df_names_titres = df_names_titres.drop(columns=['nconst','birthYear','deathYear','primaryProfession'])
df_names_titres = df_names_titres.assign(knownForTitles=df_names_titres['knownForTitles'].str.split(','))
df_names_titres = df_names_titres.explode('knownForTitles')
df_names_titres.head(15)
```

```python
df_names_titres[df_names_titres['primaryName']=='Vaughn Taylor']
```

```python
df_names_titres.info()
```

```python
df_merge_1 = df_top_5_actors_per_periods.merge(df_names_titres, how='inner', on='primaryName')
df_merge_1.head(20)
```

```python
df_merge_1 = df_merge_1[df_merge_1['knownForTitles']!='\\N']
```

```python
df_merge_1['knownForTitles'].unique()
```

```python
df_merge_1
```

```python
title_basics = pd.read_csv(chemin_bd + 'title.basics.tsv', sep='\t')
```

```python
tb = title_basics.copy()
tb = tb[['tconst','primaryTitle']]
tb.head()
```

```python
df_top5_act_films = df_merge_1.merge(tb, how='inner', left_on='knownForTitles', right_on='tconst')
df_top5_act_films
```

```python
df_top5_act_films = df_top5_act_films.groupby(['periode','primaryName','count'], sort=False)['primaryTitle'].agg(list).reset_index()
df_top5_act_films['primaryTitle']=df_top5_act_films['primaryTitle'].apply(lambda x : str(x).replace('[','').replace(']',''))
df_top5_act_films['primaryTitle']=df_top5_act_films['primaryTitle'].apply(lambda x : str(x).replace("'",'').replace("'",''))
df_top5_act_films
```

```python
df_top5_act_films.to_csv(chemin_bd + '/resultat/df_top5_act_films.csv', index=False)
```

```python
# MOYENNE DE COUNT
```

```python
df_top5_mean = df_top_5_actors_per_periods.copy()
df_top5_mean = df_top5_mean.drop(columns='primaryName')
df_top5_mean = df_top5_mean.groupby('periode', sort=False)['count'].mean().round(0).apply(lambda x : int(x)).reset_index()
```

```python
df_top5_mean.to_csv(chemin_bd + '/resultat/df_top5_mean.csv', index=False)
```

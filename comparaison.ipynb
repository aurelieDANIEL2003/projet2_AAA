```python
df_ratings= pd.read_csv(chemin_bd + 'title.ratings.tsv', sep='\t', na_values='\\N')
```

```python
df_name_nba31.head()
df_merged = pd.merge(df_tb2a, df_ratings, on="tconst", how="inner")
df_merged = df_merged[df_merged['titleType'] == 'movie']
df_merged.reset_index(drop=True, inplace=True)
df_merged.head()
```

```python
# Pour supprimer uniquement les lignes où la colonne startYear contient des NaN :
df_merged = df_merged.dropna(subset=['startYear'])
```

```python
df_merged['runtimeMinutes'] = pd.to_numeric(df_merged['runtimeMinutes'], errors='coerce')
# Pour supprimer uniquement les lignes où la colonne runtimeMinutes contient des NaN :
df_merged = df_merged.dropna(subset=['runtimeMinutes'])
```

```python
#Définir les intervalles pour les périodes
bins = [0, 1950, 1970, 1990, 2000, 2010, float('inf')]
labels = ['<1950', '1950-1970', '1970-1990', '1990-2000', '2000-2010', '>2010']
df_merged['période'] = pd.cut(df_merged['startYear'], bins=bins, labels=labels, right=False)# Ajouter une colonne 'periode' directement
```

```python
top_film_mieux_notés = df_merged.sort_values(by="averageRating", ascending=False)
top_film_mieux_notés1 = top_film_mieux_notés.reset_index(drop=True)
top_film_mieux_notés1  #.shape
#originalTitle
```

```python
result = (
    top_film_mieux_notés1.sort_values(by=["averageRating", "runtimeMinutes"], ascending=False) # Trier par note décroissante
    .groupby("période")  # Grouper par période
    .apply(lambda x: ", ".join(x.head(3)["originalTitle"]))  # Combiner les 3 meilleurs films par période
    .reset_index(name="Top 3 films")  # Réinitialiser l'index avec une nouvelle colonne pour les résultats
)
```

```python
result
```

```python
# renomer le fichier CSV
result.to_csv(chemin_bd+"resultat/resulta.csv", index = False)
```

```python
# Lire le fichier CSV
df_merged11 = pd.read_csv(chemin_bd + "df_merged1.csv")
```

```python
bins = [0, 1950, 1970, 1990, 2000, 2010, float('inf')]
labels = ['<1950', '1950-1970', '1970-1990', '1990-2000', '2000-2010', '>2010']

# Ajouter une colonne 'période' en fonction de la colonne startYear
df_merged11['période'] = pd.cut(df_merged11['startYear'], bins=bins, labels=labels, right=False)

# Calculer l'âge au moment de startYear
df_merged11['age'] = df_merged11['startYear'] - df_merged11['birthYear']

# Mettre NaN pour l'âge si l'acteur est mort avant startYear
df_merged11.loc[
    (df_merged11['deathYear'].notna()) & (df_merged11['deathYear'] < df_merged11['startYear']),
    'age'
] = np.nan

# Calculer l'âge moyen par période
age_moyen_par_periode = df_merged11.groupby('période')['age'].mean().reset_index()

age_moyen_par_periode['age'] = age_moyen_par_periode['age'].round()

age_moyen_par_periode['age'] = age_moyen_par_periode['age'].apply(lambda x : int(x))

age_moyen_par_periode
```

```python
age_moyen_par_periode.to_csv(chemin_bd+"resultat/age_moyen.csv", index = False)
```

```python
df_merge_actor = df_merge2_gros.drop(columns=['Unnamed: 0', 'tconst', 'runtimeMinutes', 'primaryProfession', 'nconst'])
```

```python
# on va voir le pourcentage de valeurs nulles par rapport au nombre de valeurs
# 1. Compter les valeurs nulles dans la colonne startYear
df_merge_actor['startYear'].isnull().sum()
```

```python
# on convertit en numerique
df_merge_actor['startYear'] = pd.to_numeric(df_merge_actor['startYear'], errors='coerce')
```

```python
# Définir les bins (intervalles) et labels (périodes)
bins = [0, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2100]
labels = ['<1900', '1900-1910', '1910-1920', '1920-1930', '1930-1940', '1940-1950', '1950-1960',
          '1960-1970', '1970-1980', '1980-1990', '1990-2000', '2000-2010', '2010-2020', '2020+']

# Ajouter une colonne 'periode' directement
df_merge_actor['periode'] = pd.cut(df_merge_actor['startYear'], bins=bins, labels=labels, right=False)

# Identifier les acteurs les plus présents pour chaque période
# actors_by_period = df_merge_actor.groupby('periode')['primaryName'].value_counts().reset_index(name='count')

```

```python
# on copie le df
df_merge_actorf = df_merge_actor.copy()
```

```python
# on garde que les films
df_merge_actorf = df_merge_actorf[df_merge_actorf['titleType'].isin(['movie'])] 
```

```python
# Identifier les 5 acteurs ou actrices les plus présents pour chaque période
top_5_actors_per_periodf = (
    df_merge_actorf.groupby('periode')['primaryName']
    .value_counts()
    .groupby(level=0, group_keys=False)
    .head(5)
    .reset_index(name='count'))
```

```python
top_5_actors_per_periodf.to_csv(chemin_bd + 'resultat/df_top_5_actors_per_periodfa.csv', index = False)
```

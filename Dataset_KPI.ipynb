```python
df_top_5_actors_per_periods = pd.read_csv(chemin_bd + 'resultat/df_top_5_actors_per_periodsa.csv')
df_top_5_actors_per_periods
```

```python
df_top_5_actors_per_periods = df_top_5_actors_per_periods.rename(columns={'periode': 'période'})
df_top_5_actors_per_periods
```

```python
#df_grouped = df.groupby('identifiant', as_index=False).agg({'nom': list})
#ch = ch.replace('[','').replace(']','')
df_top_5_actors_per_periods  = df_top_5_actors_per_periods.groupby('période',sort=False,as_index=False).agg({'primaryName': list,'count': list})
df_top_5_actors_per_periods
```

```python
df_top_5_actors_per_periods['primaryName']=df_top_5_actors_per_periods['primaryName'].apply(lambda x : str(x).replace('[','').replace(']',''))
df_top_5_actors_per_periods['count']=df_top_5_actors_per_periods['count'].apply(lambda x : str(x).replace('[','').replace(']',''))
df_top_5_actors_per_periods
```

```python
df_top_5_actors_per_periods['primaryName']=df_top_5_actors_per_periods['primaryName'].apply(lambda x : str(x).replace("'",'').replace("'",''))
df_top_5_actors_per_periods
```

```python
import streamlit as st
import pandas as pd

# Charger le fichier CSV
chemin_bd = r"./bd_ignore/"  # Remplacez par le chemin correct
df_resulta = pd.read_csv(chemin_bd + 'resultat/resulta.csv')

# Titre de la section
st.title("Top 3 films par période")

# Afficher les informations du DataFrame dans Streamlit
for index, row in df_resulta.iterrows():
    st.subheader(f"Période : {row['période']}")
    st.write(f"**Top 3 films** : {row['Top 3 films']}")
```

```python
import pandas as pd
import numpy as np
```

```python
chemin_bd = r"./bd_ignore/"
```

```python
df_tmdb = pd.read_csv(chemin_bd + 'tmdb_full.csv') 
```

```python
df_tmdb['release_date'] = pd.to_datetime(df_tmdb['release_date'], errors='coerce')
df_tmdb['year'] = df_tmdb['release_date'].dt.year
```

```python
df_tmdb.head()
```

```python
df_tmdb = df_tmdb[df_tmdb['vote_count']>100]
```

```python
df_best_movies = df_tmdb.sort_values('vote_average', ascending=False)
```

```python
df_best_movies.head(2)
```

```python
#Définir les intervalles pour les périodes
bins = [0, 1950, 1970, 1990, 2000, 2010, float('inf')]
labels = ['<1950', '1950-1970', '1970-1990', '1990-2000', '2000-2010', '>2010']
df_best_movies['période'] = pd.cut(df_best_movies['year'], bins=bins, labels=labels, right=False)# Ajouter une colonne 'periode' directement
```

```python
df_best_movies.columns
```

```python
df_best_movies = df_best_movies.drop(columns=['adult', 'backdrop_path', 'homepage', 'id', 'overview', 'poster_path','revenue', 'spoken_languages', 'status', 'tagline', 'video', 'budget', 'production_companies_country', 'popularity', 'vote_average', 'vote_count', 'popularity', 'original_title'])
```

```python
# filtrer les genres
df_best_movies = df_best_movies[df_best_movies['genres'].str.contains("|".join(['Action', 'Drama', 'Comedy', 'Horror', 'Family', 'Adventure', 'Animation', 'Thriller']))]
```

```python
df_best_movies['year'] = df_best_movies['year'].apply(lambda x: int(x))
```

```python
df_best_movies.head()
```

```python
def format(serie):
    #Enlever les []
    serie=serie.apply(lambda x : str(x).replace('[','').replace(']',''))
    serie=serie.apply(lambda x : str(x).replace("'",'').replace("'",''))
    return serie
```

```python
df_best_movies['genres'] = format(df_best_movies['genres'])
df_best_movies['production_countries'] = format(df_best_movies['production_countries'])
df_best_movies['production_companies_name'] = format(df_best_movies['production_companies_name'])
```

```python
df_best_movies
```

```python
best_movies_map = df_best_movies.copy()
```

```python
df_best_movies = df_best_movies.set_index('title')
```

```python
df_best_movies
```

```python
df_best_movies = df_best_movies.transpose()
```

```python
df_best_movies.to_csv('df_best_movies.csv', index_label='caractéritiques')
```
